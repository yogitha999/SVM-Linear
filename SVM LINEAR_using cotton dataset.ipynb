```python
X = traindata.iloc[:,0:4]
Y = traindata.iloc[:,2]
C = testdata.iloc[:,2]
T = testdata.iloc[:,0:4]
```

```python
scaler = Normalizer().fit(X)
```

```python
trainX = scaler.transform(X)
```

```python
scaler = Normalizer().fit(X)
trainX = scaler.transform(X)
```

```python
scaler = Normalizer().fit(T)
testT = scaler.transform(T)
```

```python
traindata = np.array(trainX)
trainlabel = np.array(Y)
```

```python
testdata = np.array(testT)
testlabel = np.array(C)
```

```python
from sklearn import svm
C_2d_range = [1e-2, 1, 1e2]
gamma_2d_range = [1e-1, 1, 1e1]
```

```python
for C in C_2d_range:
    for gamma in gamma_2d_range:
      print("--------------------------------------------------")
      print("c")
      print(C)
      print("gamma")
      print(gamma)
```

```python
model = svm.SVC(kernel='linear', C=100, gamma=1,probability=True) 
model = model.fit(traindata, trainlabel)
```

```python
expected = testlabel
predicted = model.predict(testdata)
```

```python
prob = model.predict_proba(testdata)
np.savetxt('svmpredicted.txt',prob)
print("--------------------------------------SVM----linear----------------------------------")
y_train1 = expected
y_pred = predicted
accuracy = accuracy_score(y_train1, y_pred)
recall = recall_score(y_train1, y_pred , average="micro")
precision = precision_score(y_train1, y_pred , average="micro")
f1 = f1_score(y_train1, y_pred, average="macro")
```
